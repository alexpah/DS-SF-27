```python
smf.ols(formula = 'ScaledSoundPressureLevel ~ AngleOfAttack + ChordLength + FreeStreamVelocity + SuctionSideDisplacementThickness + Frequency', data = df).fit().summary()
```

```python
model = smf.ols(formula = 'AngleOfAttack ~ ChordLength + FreeStreamVelocity + SuctionSideDisplacementThickness + Frequency + ScaledSoundPressureLevel ', data = df).fit()
model.summary()
```

```python
sm.qqplot(model.resid, line = 's')
```
